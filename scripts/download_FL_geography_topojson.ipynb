# Download Florida Geography Data

Downloads Census Bureau TIGER/Line Cartographic Boundary shapefiles for Florida and converts them directly to TopoJSON.

**Requirements:**
```bash
pip install geopandas topojson
```


````python
import geopandas as gpd
import topojson as tp
from pathlib import Path

# Output directory (public folder)
PUBLIC_DIR = (Path.cwd().parent / "public").resolve()
print(f"Output directory: {PUBLIC_DIR}")

# Florida FIPS code
STATE_FIPS = "12"
````

````python
# Census Bureau Cartographic Boundary URLs (2023, 500k resolution)
URLS = {
    "block_group": f"https://www2.census.gov/geo/tiger/GENZ2023/shp/cb_2023_{STATE_FIPS}_bg_500k.zip",
    "tract": f"https://www2.census.gov/geo/tiger/GENZ2023/shp/cb_2023_{STATE_FIPS}_tract_500k.zip",
    "county": "https://www2.census.gov/geo/tiger/GENZ2023/shp/cb_2023_us_county_500k.zip"
}

print("Will download from:")
for name, url in URLS.items():
    print(f"  {name}: {url}")
````

````python
# Download and convert Block Groups
print("Downloading block groups...")
gdf_bg = gpd.read_file(URLS["block_group"])
print(f"  Downloaded {len(gdf_bg)} block groups")
print(f"  Columns: {list(gdf_bg.columns)}")

# Convert to TopoJSON
topo_bg = tp.Topology(gdf_bg, prequantize=True, object_name="block_group")
output_path = PUBLIC_DIR / "FL_block_group_geo.topojson"
with open(output_path, "w") as f:
    f.write(topo_bg.to_json())
print(f"  Saved to {output_path}")
print(f"  Size: {output_path.stat().st_size / 1024 / 1024:.2f} MB")
````

````python
# Download and convert Tracts
print("Downloading tracts...")
gdf_tract = gpd.read_file(URLS["tract"])
print(f"  Downloaded {len(gdf_tract)} tracts")
print(f"  Columns: {list(gdf_tract.columns)}")

# Convert to TopoJSON
topo_tract = tp.Topology(gdf_tract, prequantize=True, object_name="tract")
output_path = PUBLIC_DIR / "FL_tract_geo.topojson"
with open(output_path, "w") as f:
    f.write(topo_tract.to_json())
print(f"  Saved to {output_path}")
print(f"  Size: {output_path.stat().st_size / 1024 / 1024:.2f} MB")
````

````python
# Download and convert Counties (filter to Florida)
print("Downloading counties (US-wide, then filtering to Florida)...")
gdf_county_us = gpd.read_file(URLS["county"])
gdf_county = gdf_county_us[gdf_county_us["STATEFP"] == STATE_FIPS].copy()
print(f"  Downloaded {len(gdf_county)} Florida counties")
print(f"  Columns: {list(gdf_county.columns)}")

# Convert to TopoJSON
topo_county = tp.Topology(gdf_county, prequantize=True, object_name="county")
output_path = PUBLIC_DIR / "FL_county_geo.topojson"
with open(output_path, "w") as f:
    f.write(topo_county.to_json())
print(f"  Saved to {output_path}")
print(f"  Size: {output_path.stat().st_size / 1024 / 1024:.2f} MB")
````

````python
print("\n✅ Done! Created:")
print(f"  - FL_block_group_geo.topojson")
print(f"  - FL_tract_geo.topojson") 
print(f"  - FL_county_geo.topojson")
````
